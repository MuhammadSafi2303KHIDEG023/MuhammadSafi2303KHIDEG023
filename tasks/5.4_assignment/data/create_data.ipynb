```python
def create_new_dataset(df,no_of_datasets,file_name_prefix,date,df_file_name):
    # no_of_datasets for amounts of data sets you want to create.
    for datasets_to_make in range(no_of_datasets): 
        # generating random numbers from range
        rand_values= [np.random.randint(df['earnings'].min(),df['earnings'].max()) for x in range(1,len(df['earnings']))]
        
        new_df=df.copy()
        new_df['earnings']= new_df['earnings'].replace([new_df['earnings']],[rand_values])
       
        # Automatically generating the folders and adding the files to it

        os.makedirs('output_data/employee_earnings/'+file_name_prefix+'-'+str(int(date)+int(datasets_to_make)),exist_ok=True)
        new_df.to_parquet('output_data/employee_earnings/'+file_name_prefix+'-'+str(int(date)+int(datasets_to_make))+'/'+df_file_name+'.parquet')
    

    
    
```

```python
create_new_dataset(earnings_df_1,2,'earnings_date=2022-02','15','employee_earnings')
```
