```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data
y = iris.target
```

# BEFORE PCA



```python
k_values = []
intertia_scores = []
for k in range(2,15):
    model = KMeans(n_clusters=k)
    model.fit(x)
    intertia_scores.append(model.inertia_)
    k_values.append(k)
    
module_of_second_derivative = np.abs(np.diff(np.diff(intertia_scores)))
```

## Use of value K

```python

plt.plot(k_values, intertia_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
```

```python
model = KMeans(n_clusters=3, n_init=1, max_iter=150)
model.fit(x)

all_predictions = model.predict(x)
centroids = model.cluster_centers_
```

```python
plt.scatter(x[:,0], x[:,1], c=all_predictions)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', color="black")
plt.show
```

# AFTER PCA

```python
pca = PCA().fit(x)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x)
```

```python
model = KMeans(n_clusters=3, n_init=1, max_iter=150)
model.fit(x_reduced)

all_predictions = model.predict(x_reduced)
centroids = model.cluster_centers_
```

```python
plt.scatter(x_reduced[:,0], x_reduced[:,1], c=all_predictions)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', color="black")
plt.show
```

## RESULT:
